# Load & Preprocess Data

```python
class DatasetProcessor:
    """Handles dataset loading and preprocessing for LLM fine-tuning."""
    def __init__(self, dataset_path="/kaggle/input/datasets/raseluddin/bengali-empathetic-conversations-corpus/BengaliEmpatheticConversationsCorpus .csv", split_ratio=0.9):
        self.dataset_path = dataset_path
        self.split_ratio = split_ratio
        self.dataset = None
        self.train_dataset = None
        self.val_dataset = None

    
    def load_and_preprocess(self):
        from datasets import Dataset
        # Load CSV
        df = pd.read_csv(self.dataset_path)

        # Clean if needed (basic)
        df = df.dropna(subset=['Questions', 'Answers'])
        df['Questions'] = df['Questions'].str.strip()
        df['Answers'] = df['Answers'].str.strip()

        # Format as instruction-style full sequence (no truncation)
        # We use a clear prompt format suitable for LLaMA-3.1-Instruct
        def format_example(row):
            prompt = f"""<|begin_of_text|><|start_header_id|>user<|end_header_id>
            {row['Questions']}<|eot_id|><|start_header_id|>assistant<|end_header_id>
            {row['Answers']}<|eot_id|>"""
            return {"text": prompt}

        df = df.apply(format_example, axis=1, result_type='expand')

        
        dataset = Dataset.from_pandas(df[['text']])


        # Train/val split
        split_idx = int(len(dataset) * self.split_ratio)
        self.train_dataset = dataset.select(range(split_idx))
        self.val_dataset = dataset.select(range(split_idx, len(dataset)))

        print(f"Train samples: {len(self.train_dataset)}, Val samples: {len(self.val_dataset)}")
        return self.train_dataset, self.val_dataset
```

# Evaluator

```python
import evaluate

class Evaluator:
    """Handles evaluation: perplexity, BLEU, ROUGE, and human eval."""
    def __init__(self, model, tokenizer, val_dataset):
        self.model = model
        self.tokenizer = tokenizer
        self.val_dataset = val_dataset
        self.bleu = evaluate.load("bleu")
        self.rouge = evaluate.load("rouge")
    
    def compute_perplexity(self):
        # Perplexity on val set (full sequences)
        encodings = self.tokenizer("\n\n".join([ex["text"] for ex in self.val_dataset]), return_tensors="pt")
        seq_len = encodings.input_ids.size(1)
        nlls = []
        for begin_loc in range(0, seq_len, 2048):  # Sliding window for long seq
            end_loc = min(begin_loc + 2048, seq_len)
            input_ids = encodings.input_ids[:, begin_loc:end_loc].to("cuda")
            target_ids = input_ids.clone()
            target_ids[:, :-1] = -100
            with torch.no_grad():
                outputs = self.model(input_ids, labels=target_ids)
                nlls.append(outputs.loss)
        ppl = torch.exp(torch.stack(nlls).mean())
        return ppl.item()
    
    def compute_bleu_rouge(self, generated_responses, references):
        bleu_score = self.bleu.compute(predictions=generated_responses, references=references)
        rouge_score = self.rouge.compute(predictions=generated_responses, references=references)
        return bleu_score['bleu'], rouge_score['rougeL']
    
    def generate_responses(self, inputs):
        input_len = [len(inputs[x]) for x in range(len(inputs))]
        
        inputs = self.tokenizer(inputs, return_tensors="pt", padding=True).to("cuda")
        outputs = self.model.generate(**inputs, max_new_tokens=512, do_sample=True)
        responses = [self.tokenizer.decode(out, skip_special_tokens=True) for out in outputs]
        
        generated=[]

        for x in range(len(responses)):
            generated_value = responses[x][input_len[x]+1:len(responses[x])]
            generated.append(generated_value)
        
        return responses,generated

    
    def human_evaluation(self, samples=10):
        # Manual scoring: Print samples and score empathy (1-5)
        scores = []
        for i in range(samples):
            input_text = self.val_dataset[i]['text'].split("<|start_header_id|>assistant<|end_header_id>")[0]
            response = self.generate_responses([input_text])[0]
            print(f"Input: {input_text}\nResponse: {response}")
            # Simulate human score (in real, input manually)
            score = int(input("Rate your score: "))  # Placeholder; replace with actual eval
            
            scores.append(score)
            score=0
        return sum(scores) / len(scores)
```

# UnslothStrategy

```python
import unsloth
from unsloth import FastLanguageModel
from trl import SFTTrainer,SFTConfig	

class FineTuneUnslothStrategy():
    """Unsloth for efficient fine-tuning."""
    def prepare_model(self, model_name):
        model, tokenizer = FastLanguageModel.from_pretrained(
            model_name,
            max_seq_length=8192,  # Full sequence, no reduction
            dtype=torch.bfloat16,  # Mixed precision
            load_in_4bit=True  # Quantization for memory
        )
        model = FastLanguageModel.get_peft_model(
            model,
            r=16,  # LoRA rank
            target_modules=["q_proj", "k_proj", "v_proj", "o_proj","gate_proj", "up_proj", "down_proj"],  # Attention layers
            lora_alpha=16,
            lora_dropout=0,
            bias = "none",
            random_state = 3407,
            use_gradient_checkpointing='unsloth',  # Resource opt
            use_rslora = False,  # We support rank stabilized LoRA
            loftq_config = None,
        )

        return model, tokenizer
    
    def train(self, model, tokenizer, train_dataset, val_dataset):
        trainer = SFTTrainer(
            model=model,
            tokenizer=tokenizer,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
            dataset_text_field="text",
            max_seq_length=8192,
            args = SFTConfig(
                per_device_train_batch_size = 2,
                gradient_accumulation_steps = 8,
                warmup_steps = 5,
                #num_train_epochs = 1, # Set this for 1 full training run.
                max_steps = 900,
                learning_rate = 5e-5,
                logging_steps = 10,
                optim = "adamw_8bit",
                weight_decay = 0.01,
                lr_scheduler_type = "linear",
                seed = 3407,
                output_dir = "outputs",
                report_to = "none",  # Use TrackIO/WandB etc
                eval_strategy="steps",
                eval_steps=50,
                save_strategy="steps",
                save_steps=50,             # more frequent logs
                do_eval=True,                         
             ),
        )
        trainer.can_return_loss = True
        trainer.train()
        return trainer
```

# FineTuner

```python
class LLAMAFineTuner:
    """Main class for fine-tuning, using Strategy pattern."""
    def __init__(self, model_name="meta-llama/Meta-Llama-3.1-8B-Instruct"):
        self.model_name = model_name
        self.strategy = FineTuneUnslothStrategy()
        self.model = None
        self.tokenizer = None
        self.trainer = None
        self.experiment_id = datetime.now().strftime("%Y%m%d_%H%M%S")
    
    def fine_tune(self, train_dataset, val_dataset):
        self.model, self.tokenizer = self.strategy.prepare_model(self.model_name)
        self.trainer = self.strategy.train(self.model, self.tokenizer, train_dataset, val_dataset)
        
        # Logging
        self.log_experiment()
    
    def log_experiment(self):
        import json
        logs = {
            "id": self.experiment_id,
            "model_name": self.model_name,
            "lora_config": "r=16, alpha=32, targets=attention",
            "train_loss": None,
            "val_loss": None,
            "metrics": "placeholder",
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
    
        if hasattr(self.trainer.state, 'log_history') and self.trainer.state.log_history:
            history = self.trainer.state.log_history
            
            # Find last training loss
            for entry in reversed(history):
                if 'loss' in entry:
                    logs["train_loss"] = round(entry['loss'], 4)
                    break
            
            # Find last eval loss (may be earlier than the final entry)
            for entry in reversed(history):
                if 'eval_loss' in entry:
                    logs["val_loss"] = round(entry['eval_loss'], 4)
                    break
    
        # Optional: also grab other metrics if present
        # logs["metrics"] = json.dumps({...})
    
        pd.DataFrame([logs]).to_csv("LLAMAExperiments.csv", mode='a', header=not os.path.exists("LLAMAExperiments.csv"))
    
    def log_responses(self, inputs, responses):
        logs = []
        for inp, resp in zip(inputs, responses):
            logs.append({
                "id": len(logs) + 1,
                "experiment_id": self.experiment_id,
                "input_text": inp,
                "response_text": resp,
                "timestamp": datetime.now()
            })
        return pd.DataFrame(logs).to_csv("GeneratedResponses.csv", mode='a', header=not os.path.exists("GeneratedResponses.csv"))
```

# PipeLines

```python
# Step 1: Preprocess Dataset
processor = DatasetProcessor()
train_ds, val_ds = processor.load_and_preprocess()
```

```python
# Step 2: Fine-Tune (using Unsloth by default; swap to "lora" if needed)
tuner = LLAMAFineTuner()
tuner.fine_tune(train_ds, val_ds)
```

```python
# Step 3: Evaluate
evaluator = Evaluator(tuner.model, tuner.tokenizer, val_ds)
```

# Evalution Metrics

```python
perplexity = evaluator.compute_perplexity()
```

```python
# Generate responses for BLEU/ROUGE
separator = "<|start_header_id|>assistant<|end_header_id>\n"
user_sep = "<|begin_of_text|><|start_header_id|>user<|end_header_id>\n"
eot = "<|eot_id|>"
test_inputs = [ex['text'].split(separator)[0].replace(user_sep,"").replace(eot,"").strip() for ex in val_ds.select(range(10))]
references = [ex['text'].split(separator)[1].replace(eot,"").strip() for ex in val_ds.select(range(10))]
```

```python
response,generated = evaluator.generate_responses(test_inputs)
```

```python
bleu,rouge = evaluator.compute_bleu_rouge(generated, references)
```

```python
human_score = evaluator.human_evaluation(samples=10)
```

```python
def chat_with_model():
    print("Type your message in Bengali (or 'exit' to quit)\n")
    
    while True:
        user_input = input("You: ").strip()
        
        if user_input.lower() in ['exit', 'quit', 'q']:
            print("Goodbye!")
            break
            
        # Format as Llama-3.1-Instruct style
        prompt = f"""<|begin_of_text|><|start_header_id|>user<|end_header_id>

        {user_input}<|eot_id|><|start_header_id|>assistant<|end_header_id>
        """
        
        inputs = evaluator.tokenizer(prompt, return_tensors="pt").to("cuda")
        
        output = evaluator.model.generate(
            **inputs,
            max_new_tokens=1024,
            do_sample=True,
            temperature=0.85,
            top_p=0.95,
            pad_token_id=evaluator.tokenizer.eos_token_id,
        )
        
        full_response = evaluator.tokenizer.decode(output[0], skip_special_tokens=True)
        generated = full_response[len(prompt):].strip()
        
        print("Model:", generated)
        print("-" * 60)
```
